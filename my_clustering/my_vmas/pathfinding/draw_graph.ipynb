```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import numpy as np

# CSV 파일 로드 및 리스트 파싱
df = pd.read_csv('./data/ortools_data_output.csv')
df['targets_per_agent'] = df['targets_per_agent'].apply(ast.literal_eval)
df['locations'] = df['locations'].apply(ast.literal_eval)

# n_targets 별 평균 및 표준편차 계산
grouped = df.groupby('n_targets')['time']
mean_times = grouped.mean()
std_times = grouped.std()  # 또는 .sem() 으로 표준오차

# 막대그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(mean_times.index, mean_times.values, yerr=std_times.values, 
        capsize=5, alpha=0.7, color='skyblue', label='Average Execution Time')

plt.title('Average Execution Time per Number of Targets (with Std Dev)')
plt.xlabel('Number of Targets (n_targets)')
plt.ylabel('Execution Time (seconds)')
plt.grid(True, axis='y')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns

# Clust 데이터 로드
df_clust = pd.read_csv('./data/clust_data_output.csv')
df_clust['targets_per_agent'] = df_clust['targets_per_agent'].apply(ast.literal_eval)
df_clust['locations'] = df_clust['locations'].apply(ast.literal_eval)
df_clust['method'] = 'Clust + TSP'  # 방법 이름 추가

# Ortools 데이터 로드
df_ortools = pd.read_csv('./data/ortools_data_output.csv')
df_ortools['targets_per_agent'] = df_ortools['targets_per_agent'].apply(ast.literal_eval)
df_ortools['locations'] = df_ortools['locations'].apply(ast.literal_eval)
df_ortools['method'] = 'Ortools'  # 방법 이름 추가

# 두 데이터 합치기
df_all = pd.concat([df_clust, df_ortools], ignore_index=True)

# 박스플롯 그리기
plt.figure(figsize=(12, 6))
sns.boxplot(x='n_targets', y='time', hue='method', data=df_all, palette='Set2')

plt.title('Execution Time Comparison by Number of Targets')
plt.xlabel('Number of Targets (n_targets)')
plt.ylabel('Execution Time (seconds)')
plt.grid(True, axis='y')
plt.legend(title='Method')
plt.tight_layout()
plt.show()
```
